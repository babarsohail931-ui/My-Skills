```python
sns.catplot(x="Family_Income",y="Exam_Score",data=df,kind="bar",hue="Gender",legend=True)
plt.title("Family_Income: Mean Exam_Score by Gender")
plt.xlabel("Family_Income")
plt.ylabel("Mean Exam_Score")
plt.show()
```

```python
sns.violinplot(x="Family_Income",y="Exam_Score",data=df)
plt.title("Exam_Score Distribution by Family_Income")
plt.xlabel("School_Type")
plt.ylabel("Peer_Influence")
plt.show()
```

```python
sns.boxplot(x="Family_Income",y="Exam_Score",data=df,palette="Set2")
plt.title("Exam_Score Distribution by Family_Income")
plt.xlabel("School_Type")
plt.ylabel("Peer_Influence")
plt.show()
```

```python
sns.stripplot(x="Family_Income",y="Exam_Score",data=df,jitter=True,alpha=0.6,palette="Dark2",hue="Gender")
plt.title("Stripplot mapping individual Exam_Score by Gender")
plt.xlabel("Family_Income")
plt.ylabel("Exam_Score")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f",linewidths=0.5)
plt.title("Correlation Heatmap of Family Income")
plt.show()
```
